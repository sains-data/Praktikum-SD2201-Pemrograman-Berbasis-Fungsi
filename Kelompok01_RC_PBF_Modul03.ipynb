```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_excel('TabelSumut.xlsx')

# Melakukan explorasi data
print(data.head())  # Menampilkan lima baris pertama
print(data.info())  # Informasi tentang dataset
print(data.describe())  # Statistik deskriptif
```

```python
data.dropna()  # Menghapus baris dengan nilai yang hilang
data.fillna(0)  # Mengganti nilai yang hilang dengan 0
```

```python
# Membersihkan data dari nilai kosong

datacleaned = data[(data != 0).all(axis=1)].dropna()
print(datacleaned)
```

```python
#Visualisasi
# Barplot
plt.figure(figsize=(8,6))
sns.barplot(x='Kabupaten Kota', y='Produksi (ton) 2020', data=datacleaned)
plt.title('Barplot Hasil Panen disetiap Kabupaten di Sumatera Utara')
plt.xlabel('Provinsi')
plt.ylabel('Nilai')
plt.xticks(rotation=60)
plt.show()
```

```python
# Membuat histogram
plt.figure(figsize=(8, 6))
sns.histplot(datacleaned['Produksi (ton) 2020'], bins=20)
plt.title('Histogram Produksi pada tahun 2020')
plt.xlabel('Produksi (ton) 2020')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Kabupaten Kota', y='Produksi (ton) 2020', data=datacleaned)
plt.title('Boxplot')
plt.xlabel('Kategori')
plt.ylabel('Nilai')
plt.xticks(rotation=60)
plt.show()
```

```python
# Membuat scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Kabupaten Kota', y='Produksi (ton) 2020', data=datacleaned)
plt.title('Scatter Plot')
plt.xlabel('Kabupaten Kota')
plt.ylabel('Produksi pada tahun 2020')
plt.xticks(rotation=60)
plt.show()
```

```python
# Menghitung median, mean, dan modus untuk setiap kolom
median_values = datacleaned.median()
mean_values = datacleaned.mean()
mode_values = datacleaned.mode().iloc[0]

# Menampilkan hasil
print("Median:")
print(median_values)
print("\nMean:")
print(mean_values)
print("\nModus:")
print(mode_values)
```
