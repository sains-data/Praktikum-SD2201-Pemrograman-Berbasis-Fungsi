```python
from functools import reduce

class OnlineShoppingApp:
    def __init__(self):
        self.barang = {"kemeja": 199999, "celanapanjang":249000, "kaos": 149000, "celanapendek": 99000, "kaoskaki": 89000}
        self.stok = {"kemeja": 1000, "celanapanjang": 1000, "kaos": 2000, "celanapendek": 1000, "kaoskaki": 1500}
        self.cart = []

    def display_items(self):
        print("Daftar Barang:")
        for item, harga in self.barang.items():
            print(f"{item} - Harga: {harga} - Stok: {self.stok[item]}")

    def order_item(self, item, quantity):
        if item in self.barang and self.stok[item] >= quantity:
            self.cart.append({"item": item, "quantity": quantity})
            self.stok[item] -= quantity
            print(f"{quantity} {item} berhasil ditambahkan ke dalam keranjang.")
        elif item in self.barang and self.stok[item] < quantity:
            print(f"Maaf, hanya tersedia {self.stok[item]} {item}.")
        else:
            print(f"Maaf, {item} tidak tersedia dalam daftar barang.")

    def order_items(self):
        while True:
            item = input("Masukkan nama barang yang ingin dibeli (atau ketik 'keranjang' untuk melihat keranjang atau 'selesai' untuk mengakhiri pemesanan): ").lower()
            if item == "keranjang":
                self.display_cart()
                continue
            elif item == "selesai":
                break
            elif item not in self.barang:
                print(f"Maaf, {item} tidak tersedia dalam daftar barang.")
                continue
            elif self.stok[item] == 0:
                print(f"Maaf, stok {item} habis.")
                continue
            else:
                quantity = int(input(f"Masukkan jumlah {item} yang ingin dibeli: "))
                self.order_item(item, quantity)

    def display_cart(self):
        if not self.cart:
            print("Keranjang belanja masih kosong.")
        else:
            print("Keranjang belanja:")
            for item in self.cart:
                print(f"{item['item']} - Jumlah: {item['quantity']}")

    def calculate_total_sales(self):
        # Fungsi untuk menghitung total penjualan
        calculate_total = lambda acc, item: acc + (self.barang[item['item']] * item['quantity'])

        # Menghitung total penjualan menggunakan reduce dan lambda
        total_sales = reduce(calculate_total, self.cart, 0)
        return total_sales

# Membuat objek aplikasi belanja online
online_shopping_app = OnlineShoppingApp()

# Menampilkan daftar barang
online_shopping_app.display_items()

# Pembeli memesan barang
online_shopping_app.order_items()

# Menampilkan keranjang belanja
online_shopping_app.display_cart()

# Menghitung dan menampilkan total penjualan
total_sales = online_shopping_app.calculate_total_sales()
print(f"\nTotal Harga Barang: {total_sales}")
```
