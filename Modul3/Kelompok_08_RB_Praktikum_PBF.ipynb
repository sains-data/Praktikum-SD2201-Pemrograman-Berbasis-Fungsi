```python
import numpy as np
import pandas as pd
```

```python
def pca(X, n_components):
  X_centered = X - np.mean(X, axis=0)
  U, S, Vt = np.linalg.svd(X_centered)
  return U[:, :n_components] @ np.diag(S[:n_components])
```

```python
iris_data = pd.read_csv("iris.csv")
iris_data
```

```python
X = iris_data.select_dtypes(include=['int', 'float'])
n_components = 2
X
```

```python
X_pca = pca(X, n_components)
print("Komponen Utama:\n", X_pca)
```
