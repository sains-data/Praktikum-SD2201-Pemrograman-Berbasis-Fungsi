```python
import numpy as np
import matplotlib.pyplot as plt

def time_series_generator(n, trend_slope, seasonality_period, seasonality_amplitude):
    for t in range(n):
        trend = trend_slope * t
        seasonality = seasonality_amplitude * np.sin(2 * np.pi * t / seasonality_period)
        yield trend + seasonality

# Penggunaan:
n = 100
trend_slope = 0.5
seasonality_period = 10
seasonality_amplitude = 2

time_series_data = list(time_series_generator(n, trend_slope, seasonality_period, seasonality_amplitude))

# Visualisasi data time series
plt.plot(time_series_data)
plt.title('Time Series Data with Trend and Seasonality')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
