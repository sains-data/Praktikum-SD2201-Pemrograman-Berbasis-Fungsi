```python
import numpy as np
import matplotlib.pyplot as plt

def metode_simpson(f, a, b, n):
    h = (b - a) / n
    result = f(a) + f(b)
    for i in range(1, n, 2):
        result += 4 * f(a + i * h)
    for i in range(2, n-1, 2):
        result += 2 * f(a + i * h)
    return result * h / 3

# Contoh fungsi data (misalnya, sebuah fungsi kuadrat)
def f(x):
    return x**2 + 5*np.sin(x)

# Batas bawah dan atas dari interval
a = 0
b = 5

# Jumlah subinterval (pastikan jumlahnya genap untuk metode Simpson)
n = 10

# Menghitung titik-titik x dan y pada kurva
x_val = np.linspace(a, b, n+1)
y_val= f(x_val)

# Menggunakan metode integrasi Simpson untuk menghitung luas di bawah kurva
luas_daerah_dibawah_kurva = metode_simpson(f, a, b, n)

# Plot kurva dan area di bawah kurva
plt.plot(x_val, y_val, label='Kurva $y = x^2 + 5sin(x)$')
plt.fill_between(x_val, y_val, color='lightblue', alpha=0.5, label='Luas di Bawah Kurva')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Luas Daerah Di Bawah Kurva')
plt.legend()
plt.grid(True)
plt.text(2, 10, f'Luas = {luas_daerah_dibawah_kurva:.2f}', fontsize=12, ha='center')
plt.show()
```
